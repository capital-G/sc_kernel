# *SuperCollider* kernel for Jupyter

Using Jupyter for Supercollider can help us to tell a story with the code as it is possible to write markdown code between SuperCollider code, add LaTEX code and add media assets in between.

```supercollider
(60 + (1..10)).midicps
```

We can explain our code with formulas and implement thos formulas in SuperCollider.

$$
\lbrace 2^{i/12} | i \in  \lbrace 0..12 \rbrace \rbrace
$$

As a bonus we do not have to write `( )` arounde each code block we want to execute because every Cell gets surrounded by those parenthesis automatically.

```supercollider
Array.fill(13, {|i|
    2**(i/12)
})
```
